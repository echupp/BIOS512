```r
library("tidyverse")
```

```r
turbine_data = 
    read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/master/data/2020/2020-10-27/wind-turbine.csv',
    col_types = cols()
)

head(turbine_data)
```

Each row in the table represents a single turbine.

Make a bar chart 📊 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
turbine_data = turbine_data |>
    mutate(prov_sorted = province_territory |> fct_infreq() |> fct_rev())

turbine_data |> 
    ggplot(aes(y = prov_sorted)) +
    geom_bar() +
    labs(
        x = "Count",
        y = "Province")
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
turbine_data = turbine_data |>
    mutate(man_sorted = manufacturer |> fct_lump(n=5) |> 
           fct_lump(n = 7) |> 
           fct_infreq() |> 
           fct_relevel('Other', after= Inf) |>
           fct_rev()
    )

turbine_data |> 
    ggplot(aes(y = man_sorted)) +
    geom_bar() +
    labs(
        x = "Count",
        y = "Manufacturer")
```

```r
aoptions(repr.plot.width = 4, repr.plot.height = 4, repr.plot.res = 250)

ggplot(data = canada) + 
    geom_polygon(aes(x = long, y = lat, group = group), fill = 'coral', color = "white", alpha = 0.5) + 
    coord_quickmap() +
    geom_point(data = turbine_data, mapping = aes(x = longitude, y = latitude), alpha = 0.5, color = "slateblue", size = 0.4)
```
